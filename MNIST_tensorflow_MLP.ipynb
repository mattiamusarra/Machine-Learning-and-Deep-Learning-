```python
import tensorflow as tf
from tensorflow.keras import layers, models
import datetime


mnist = tf.keras.datasets.mnist

(x_train, y_train),(x_test,y_test) = mnist.load_data()

#Normalizziamo: i pixel vanno da 0 a 255. Li portiamo tra 0 e 1 per aiutare la convergenza.
x_train, x_test = x_train / 255.0, x_test /255.0

#Funzione per configurare TensorBoard

def get_tb_callback(name):
  log_dir = 'log/mnist/' + name + "_" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
  return tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
model_mlp = models.Sequential([
    layers.Flatten(input_shape=(28,28)), #Appiattisce l'immagine
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(10, activation="softmax")
])

model_mlp.compile(optimizer = "adam",
loss="sparse_categorical_crossentropy",
metrics=['accuracy'])

print("Allenamento MLP...")
model_mlp.fit(x_train, y_train, epochs=5, validation_data=(x_test,y_test), callbacks = [get_tb_callback("MLP")])
```

```python
x_train_cnn = x_train.reshape((-1,28,28,1))
x_test_cnn = x_test.reshape((-1,28,28,1))

model_cnn = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

```python
model_cnn.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

print("\nAllenamento CNN....")
model_cnn.fit(x_train_cnn, y_train, epochs=5, validation_data=(x_test_cnn,y_test), callbacks=[get_tb_callback("CNN")])
```
