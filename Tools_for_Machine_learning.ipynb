```python
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[10,20,25,30]

plt.plot(x,y,label="Andamento Dati")
plt.title("Grafico a linee semplici")
plt.xlabel("Asse X")
plt.ylabel("Asse Y")
plt.legend()
plt.show()
```

```python
from matplotlib.lines import lineStyles
#Grafico a Barre
studenti = ['stud 1 ','Stud 2', 'Stud 3']
voti = [26,28,18]

#Creazione del grafico a barre
plt.figure(figsize=(8,6)) #Imposta la dimensione della figura
plt.bar(studenti,voti,color=['blue','green','red'])
plt.title("Voti degli studenti all\' Esame")
plt.xlabel("Studenti")
plt.ylabel("Voti")
for i,voto in enumerate(voti):
  plt.text(i, voto+0.5, str(voto), ha="center",va="bottom", fontsize=10)
plt.ylim(0,30)
plt.grid(axis='y',linestyle='--')
plt.show()
```

```python
import seaborn as sns

df = sns.load_dataset('tips')
plt.figure(figsize=(8,6))
sns.violinplot(x='day',y='total_bill',data=df, inner='point')
plt.title("Distribuzione del conto totale per gioirno della settimana")
plt.xlabel("Giorno")
plt.ylabel("Conto Totale")
plt.grid(axis='y', linestyle="--")
plt.show()
```

```python
df = sns.load_dataset('iris')

#Calcolo della matrice di correlazione
correlation_matrix = df.select_dtypes(include=['number']).corr()
plt.figure(figsize=(7,6))
sns.heatmap(
    correlation_matrix,
    annot=True, #Mostra i valori numerici della correlazione
    cmap="vlag", #Palette di colori
    fmt=".2f",
    linewidths=".5",
    cbar=True #Barra dei colori
)
plt.title("Matrice di correlazione tra le Misure del Fiore Iris")
plt.show()
```
